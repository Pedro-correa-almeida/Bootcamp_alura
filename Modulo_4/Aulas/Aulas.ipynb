## Desafio 1 -pesquisar aprendizado supervisionado, não supervisionado, por reforço e mais (forma).

O aprendizado de máquina, ou machine learning, é uma técnica que permite construir um modelo matemático capaz de gerar previsões. Porém esse aprendizado pode ser realizado de diversas formas dependendo do objetivo final, e cada dessa forma existem diferentes abordagens e, portanto, o papel do cientista de dados é entender o problema e achar a melhor forma do aprendizado de máquina e depois pensar no modelo propriamente dito, essas formas são:

> * Aprendizado Supervisionado : esse aprendizado é aquele aprendizado onde o modelo já possui, nesses dados passados, a resposta do que aconteceu, nesse tipo de aprendizado o modelo olha para os dados passados e tenta gerar uma fórmula matemática que consiga descrever a resposta do problema em função das variáveis das observações, os modelos mais comuns nesse casos são as regressões e classificações, como a regressão linear e regressão logística.<br>
> * Aprendizado Não Supervisionado : neste tipo de aprendizado existe um problema a ser resolvido mas não existe uma resposta pré definida, ele pode ser muito útil para descobrir classes sem pré defini-las. Um exemplo é o cluster que a partir de variáveis de observações ele tenta encontrar padrões e definir classes.Pode ser muito útil em dados não relacionais como classificação de textos.<br>
> * Aprendizado Por Reforço : nesse caso o objetivo é ensinar o modelo por tentativa e erro, se aproximando muito de um aprendizado real e tendo uma base no estudo da psicologia, onde muitas vezes só aprendemos tentando e errando. Esse tipo de aprendizado pode ser muito útil na tentativa de criar um modelo a fim de replicar tarefas humanas, como carros inteligentes que dirigem sozinhos e inteligências artificiais dentro dos jogos.<br>

## Desafio 2 - Pesquisar sobre classificação, regressão e outros problemas que podem ser solucionados com Machine Learning

> Regressão  : a regressão tem como objetivo prever um valor contínuo a partir dos dados, portanto ela é muito utilizada nas previsões de vendas, de custos e outros problemas que envolvam valores numéricos na resposta. Ela ser do tipo linear que tenta encontrar uma reta que ajusta os dados independentes(variáveis) e os dados dependentes que seriam a resposta a ser encontrada, além disso existem as regressões não lineares que conseguem identificar padrões que muitas vezes podem passar em branco, pois nesse casos ela tenta encontrar uma equação polinomial que possa descrever os dados, e, portanto consegue identificar curvas que correlacionam os dados.

> Classificação : a classificação é utilizada quando queremos prever a classe de algo, essa classe pode ser binária de sim e não, por exemplo prever se e-mail é spam, ou se uma compra realizada no cartão é fraude. Além disso existem classificações multiclasses onde os rótulos previstos pelo mdelo são variados. Um exemplo de uma previsão multiclasses é prever o tipo de flor a partir do comprimento e largura das pétalas.

O motivo de jogarmos fora o 'ICU' do dado X é porque ao estarmos trabalhando com aprendizado supervisionado temos que estar sempre atentos a definir previamente a variável dependente e as variáveis independentes.

## Desafio 5 - Decida quanto você vai escolher para treino e quanto para teste? Quanto as pessoas utilizam e qual o motivo?

Essa pergunta depende muito do tamanho total do dataset, pois ao separarmos uma parte para testar o modelo temos que pensar em ter uma amostra que consiga representar os dados totais da maneira mais equilibrada possível e ao mesmo tempo fazer a mesma coisa para os dados de treino. Portanto, se temos uma quantidade muito grande de dados podemos 'nos dar ao luxo' de deixar uma porcentagem menor dos dados destinados ao teste, no entanto, quando temos uma quantidade mais limitada dos dados temos o desafio de encontrar uma porcentagem ideal para separar os dados.

As pessoas normalmente utilizam entre 20% e 30% do dataset total para teste, pois na maior parte das vezes essa porcentagem é ideal para fazer com que cada dataset cumpra seu papel muito bem, porém podemos testar diferentes tamanhos a fim de observar a acurácia dos dois datasets ao mudar esse tamanho

## Desafio 6 - Olhe a documentação da LogisticRegression... o que tem de parâmetro em comum com o train_test_split? O que isso significa pra nós? 

As duas funções possuem o mesmo parâmetro 'random_state', este parâmetro é reponsável pode definir a semente que irá gerar os números aletórios afim de tomar decisões aleatoriamente. No caso da função train_test_split essa randomização tem como objetivo separar os dados de treino e teste da maneira mais aleatória possível.Na função LogisticRegression dependendo do método utilizado passado no parâmetro 'solver', a função precisa gerar randomizações a fim de realizar os métodos matemáticos.

Portanto isso nos diz que devemos tomar cuidado na hora de definir essa semente aletória, pois sempre que rodarmos essas funções o correto é definir uma semente novamente.

```python
#Criando um dataframe que indica se o paciente foi para UTI em algum momento
dados_qualquer_janela = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False).agg({'ICU':'max'})
dados_qualquer_janela = dados_qualquer_janela.rename(columns={'ICU':'ICU_any_window'})
dados_qualquer_janela.head()
```

```python
#Juntando o dataframe pra criar a coluna indicando se o paciente foi para UTI em algum momento
dados = dados.merge(dados_qualquer_janela, on='PATIENT_VISIT_IDENTIFIER')
dados.head()
```

```python
dados.describe()
```

## Super desafio : não usar backfill da janela após a pessoa ter ido pra UTI

Para não usar o backfill da janela após a pessoa ter ido para UTI vou eliminar os dados das janelas onde os pacientes estavam na UTI e só depois fazer o backfill, dessa forma já elimino os dados em que o paciente foi para UTI na primeira janela e evito que os dados que foram obtidos quando o paciente já estava na UTI interfira ao preencher os dados nulos.

```python
#Eliminando os dados onde 'ICU' é igual a 1
dados = dados.drop(index = dados.loc[dados['ICU'] == 1].index)

#Eliminado a coluna 'ICU' pois não será mais utilizada, em vez dela, tentarei prever a 'ICU_any_window'
dados = dados.drop(columns='ICU')
dados
```

```python
#Identificando as colunas contínuas
colunas_continuas = dados.iloc[:, 13:-2].columns

#Fazendo o prenchimento dos valores nulos
dados[colunas_continuas] = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index=False)[colunas_continuas].fillna(method='ffill').fillna(method='bfill')

#Eliminando os dados nulos restantes
dados = dados.dropna()
dados
```

```python
dados.describe()
```

```python
def select_window(rows):
    return rows.loc[rows['WINDOW'] == '0-2',:]
```

```python
#Aplicando a função que retona apenas a janela de quando o paciente chegou no hospital
dados_limpos = dados.groupby('PATIENT_VISIT_IDENTIFIER').apply(select_window)
dados_limpos.head()
```

```python
dados_limpos.describe()
```

```python
dados_limpos.set_index(dados_limpos.columns[0]).to_csv('../../Modulo_5/dados/dados.csv')
```

## Desafio 07 - Arranque fora as colunas que não devem ser utilizadas

```python
#Eliminando as colunas que indicam o ID e o genêro pois essas informações não serão úteis na previsão e podem atrapalhar 
dados_limpos = dados_limpos.drop(columns=['PATIENT_VISIT_IDENTIFIER','GENDER'])
dados_limpos = dados_limpos.reset_index(drop=True)
dados_limpos.head()
```

```python
#Renomeando a coluna 'ICU_any_window' para 'ICU' agora que estamos trabalhando com apenas 1 janela
dados_limpos = dados_limpos.rename(columns={'ICU_any_window':'ICU'})
dados_limpos.head()
```

```python
#Fazendo os tratamentos adequados para trasnformar a coluna 'AGE_PERCENTIL' para inteiro de uma forma que o modelo possa entender
dados_limpos['AGE_PERCENTIL'] = dados_limpos['AGE_PERCENTIL'].str.replace('th', '')
dados_limpos['AGE_PERCENTIL'] = dados_limpos['AGE_PERCENTIL'].str.replace('Above ', '')
dados_limpos['AGE_PERCENTIL'] = dados_limpos['AGE_PERCENTIL'].astype('int64')
```

```python
dados_limpos.describe()
```

```python
x = dados_limpos.loc[:,dados_limpos.describe().columns].drop(columns='ICU')
y = dados_limpos.loc[:,'ICU']
```

```python
np.random.seed(748596)
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.3)
```

```python
model_dummy = DummyClassifier()
model_dummy.fit(x_train, y_train)
y_pred_dummy = model_dummy.predict(x_test)
accuracy_score(y_test, y_pred_dummy)
```

```python
model_log = LogisticRegression(max_iter=10000)
model_log.fit(x_train, y_train)
y_pred_log = model_log.predict(x_test)
accuracy_score(y_test, y_pred_log)
```

```python
confusion_matrix(y_test, y_pred_log)
```

```python
print(classification_report(y_test, y_pred_log))
```
