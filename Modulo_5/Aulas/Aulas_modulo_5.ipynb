## Desafio 03: Pesquisar sobre F-score. 

F_score são métricas capazes de resumir em apenas um número a precisão e o recall de uma classificação, o intuito é conseguir avaliar e saber qual modelo é melhor com apenas uma métrica que generalize a acertividade do modelo em relação aos valores totais verdadeiros e aos valores totais preditos, pois muitas vezes olhar para precisão e para o recall separados não ajuda a escolher qual modelo é melhor, uma vez que um pode ter uma precisão melhor e o outro um recall melhor.

A fórmula dessa métrica depende de um parâmetro beta, quando esse parâmetro é igual a 1 a métrica é chamada f1-score, quando é igual a 2 é chamada de f2-score e assim sucessivamente, essa fórmula é:

![formula_f_score](https://wikimedia.org/api/rest_v1/media/math/render/svg/136f45612c08805f4254f63d2f2524bc25075fff)

```python
def f_score(precision,recall,beta):
    return (1 + (beta**2))*precision*recall/((beta**2*precision)+recall)
```

## F1-score

precisão = 0.5 <br>
recall = 1

```python
f_score(0.5,1,1)
```

precisão = 1 <br>
recall = 0.5

```python
f_score(1,0.5,1)
```

## F-2 Score

precisão = 0.5 <br>
recall = 1

```python
f_score(0.5,1,2)
```

```python
f_score(1,0.5,2)
```

precisão = 1 <br>
recall = 0.5

Como é possível perceber nos testes, quando usamos o f1-score as duas métricas de recall e precisão tem o mesmo peso, ao aumentarmos o beta um recall mais baixo impacta mais no resultado do que a precisão, ou seja, usar valores de beta maiores pode ser interessante se o recall for mais importante no problema do contexto

## Desafio 05: Criar uma função que gere a curva ROC.

```python
class Modelo:
    def __init__(self, Modelo, dados, n_roda=1, **kwargs):
        x = dados.drop(columns='ICU')
        y = dados['ICU']
        
        auc_list = []
        tprs = []
        self.fprs_mean = np.linspace(0, 1, 100)
        for i in range(n_roda):
            
            X_train, X_test, y_train, y_test = train_test_split(x,y, stratify=y)
            self.modelo = Modelo(**kwargs)
            self.modelo.fit(X_train, y_train)
            pred_proba = self.modelo.predict_proba(X_test)
            fpr, tpr, _ = roc_curve(y_test, pred_proba[:,1])

            interp_tpr = np.interp(self.fprs_mean, fpr, tpr)
            interp_tpr[0] = 0.0
            tprs.append(interp_tpr)
            auc_list.append(roc_auc_score(y_test, pred_proba[:,1]))
        
        self.tprs_mean = np.mean(tprs, axis=0)
        self.tprs_std = np.std(tprs, axis=0)
        self.auc_mean = np.mean(auc_list)
        self.auc_std = np.std(auc_list)
        
    def plota_roc_curve(self, ax, color):
        
        name = self.modelo
        
        sns.lineplot(self.fprs_mean, self.tprs_mean, ax=ax, label= r'ROC CURVE (AUC MEAN = %0.2f $\pm$ %0.2f)' % (self.auc_mean, 2*self.auc_std) + f'{name}', 
                     color=color, estimator=None)
        
        plt.sca(ax)
        tprs_upper = np.minimum(self.tprs_mean + self.tprs_std, 1)
        tprs_lower = np.maximum(self.tprs_mean - self.tprs_std, 0)
        plt.fill_between(self.fprs_mean, tprs_lower, tprs_upper, color=color, alpha=.05)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate', fontsize=15)
        plt.ylabel('True Positive Rate', fontsize=15)
        plt.legend(loc='lower right', fontsize=12)
        plt.title('ROC CURVE', fontsize=25)
        return ax
```

```python
dt = Modelo(DecisionTreeClassifier,dados, n_roda=50)
```

```python
log = Modelo(LogisticRegression,dados,  n_roda=50)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
dt.plota_roc_curve(ax, 'orange')
log.plota_roc_curve(ax, 'red')
plt.show()
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold , ShuffleSplit, GroupKFold
from sklearn.model_selection import cross_validate
from sklearn.metrics import plot_roc_curve
```

```python
class Modelo:
    
    def __init__(self, estimator, dados, n_splits, n_repeats, target_variable='ICU', **kwargs):
        
        random_state = np.random.RandomState(654354)
        dados = dados.sample(frac=1, random_state=random_state).reset_index(drop=True)
        self.x = dados.drop(columns=target_variable)
        self.y = dados[target_variable]
        
        self.modelo = estimator(**kwargs)
        self.cv = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats)
        resultados = cross_validate(self.modelo, self.x, self.y, cv=self.cv, scoring='roc_auc', return_train_score=True)
        self.auc_medio = np.mean(resultados['test_score'])
        self.auc_medio_treino = np.mean(resultados['train_score'])
        self.auc_std = np.std(resultados['test_score'])
        self.auc_intervalo = (self.auc_medio - self.auc_std, self.auc_medio + self.auc_std)
        print(f'AUC Médio {self.auc_medio} - {self.auc_medio_treino}')
        
    def plota_roc_curve(self, ax, color):
        
        tprs = []
        aucs = []
        fpr_mean = np.linspace(0, 1, 100)
        for i, (train, test) in enumerate(self.cv.split(self.x, self.y)):
            self.modelo.fit(self.x.iloc[train,:], self.y.iloc[train])
            pred_proba = self.modelo.predict_proba(self.x.iloc[test,:])
            fpr, tpr, _ = roc_curve(self.y.iloc[test], pred_proba[:,1])
            interp_tpr = np.interp(fpr_mean, fpr, tpr)
            interp_tpr[0] = 0.0
            tprs.append(interp_tpr)
            aucs.append(roc_auc_score(self.y[test], pred_proba[:,1]))
        
        tprs_mean = np.mean(tprs, axis=0)
        tprs_std = np.std(tprs, axis=0)
        auc_mean = np.mean(aucs)
        auc_std = np.std(aucs)

        name = self.modelo
        sns.lineplot(fpr_mean, tprs_mean, ax=ax, label= r'ROC CURVE (AUC MEAN = %0.2f $\pm$ %0.2f)' % (auc_mean, 2*auc_std) + f'{name}', 
                     color=color, estimator=None)
        
        plt.sca(ax)
        tprs_upper = np.minimum(tprs_mean + tprs_std, 1)
        tprs_lower = np.maximum(tprs_mean - tprs_std, 0)
        plt.fill_between(fpr_mean, tprs_lower, tprs_upper, color=color, alpha=.05)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate', fontsize=15)
        plt.ylabel('True Positive Rate', fontsize=15)
        plt.legend(loc='lower right', fontsize=12)
        plt.title('ROC CURVE', fontsize=25)
        
        return ax
```

## Desafio 08: Testar outros splitter classes e observar as diferenças.



O splitter ShuffleSplit não tem a preocupação de usar todos os dados para treino e para teste, o seu objetivo é rodar o modelo um número específico de vezes e cada vez com um conjunto aleatório dos dados de treino e teste.

```python
log = LogisticRegression()
cv = ShuffleSplit(n_splits=50)
resultados = cross_validate(log, x, y, cv=cv, scoring='roc_auc')
auc_medio = np.mean(resultados['test_score'])
auc_std = np.std(resultados['test_score'])
auc_intervalo = (auc_medio - auc_std, auc_medio + auc_std)
print(f'AUC Médio {auc_medio}')
print(f'Intevalo AUC {auc_intervalo}')
```

Pelo resultado ele apresentou uma média muito próxima porém um desvio padrão maior

O GroupKFold garante que um mesmo grupo não vai aparecer em diferentes folds na hora da validação cruzada, a fim apenas de testar o impacto de separarmos os dados pela idade vou utilizar esse método para separar em grupos de pessoas pela idade, ou seja, o modelo treinará e testará pessoas sempre de idades de grupos diferentes, o esperado é um desempenho abaixo.

```python
log = LogisticRegression()
groups = np.array(x['AGE_PERCENTIL'])
cv = GroupKFold(n_splits=9)
resultados = cross_validate(log, x, y, cv=cv, scoring='roc_auc', groups=groups)
auc_medio = np.mean(resultados['test_score'])
auc_std = np.std(resultados['test_score'])
auc_intervalo = (auc_medio - auc_std, auc_medio + auc_std)
print(f'AUC Médio {auc_medio}')
print(f'Intevalo AUC {auc_intervalo}')
```

Como é possível perceber quando o modelo não treina com todos os grupos de idade ele obtem uma performance bem abaixo, ou seja, temos que garantir que ele treinará com com todos os tipos de pessoas em idades diferentes, uma vez que essa informação é muito importante para prever a internação.

## Desafio 09: Deixar o gráfico de AUC bonitão

```python
def apresentacao(ax, title:str='', subtitle:str='', xlabel:str='', ylabel:str='',style='white', spines_invisible:list=['top', 'right']):

    sns.set_style(style)
    plt.sca(ax)
    plt.title(title, fontsize=25, loc='left', color='#333333')
    plt.xlabel(xlabel, color='#333333', fontsize=15)
    plt.ylabel(ylabel, color='#333333', fontsize=15)
    for spine in spines_invisible:
        ax.spines[spine].set_visible(False)
```

```python
def plota_auc(range_max_depth, kwargs_apresentacao={}, **kwargs):
    fig, ax = plt.subplots(figsize=(20,10))
    
    x = np.array(range_max_depth)
    y_test = []
    y_train = []
    for max_depth in range_max_depth:
        model = Modelo(DecisionTreeClassifier, max_depth=max_depth, **kwargs)
        y_test.append(model.auc_medio)
        y_train.append(model.auc_medio_treino)
        
    sns.lineplot(x, y_test, ax=ax, color='darkblue')
    sns.lineplot(x, y_train, ax=ax, color='red')
    
    plt.annotate('AUC Teste', (max(x)+0.005, y_test[-1]), color='darkblue', fontsize=15)
    plt.annotate('AUC Treino', (max(x)+0.005, y_train[-1]), color='red', fontsize=15)
    
    apresentacao(ax, **kwargs_apresentacao)
    plt.yticks(fontsize=15, color='#333333')
    plt.xticks(x,fontsize=15, color='#333333')
    plt.ylim([0,1.05])
```

```python
def drop_correlation(data, threshold):
    
    data_float = data.select_dtypes('float64')
    matrix_corr = data_float.corr().abs()
    matrix_corr_upper = matrix_corr.where(np.triu(np.ones(matrix_corr.shape), k=1).astype(np.bool))
    cols_drop = [col for col in matrix_corr_upper.columns if any(matrix_corr_upper[col] >= threshold)]
    return data.drop(cols_drop, axis=1)
```
