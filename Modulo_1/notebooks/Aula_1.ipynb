```python
import pandas as pd

dados = pd.read_csv("A151346189_28_143_208.csv", encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",", engine="python")
```

```python
dados.head()
```

```python
pd.options.display.float_format = "{:.2f}".format
dados.mean()
```

## Visualização

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(12,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando o gráfico 
sns.barplot(x='Unidade da Federação', y='2008/Ago', data=dados.sort_values('2008/Ago'), palette='Reds', ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title('Gasto Hospitalar Público Processado Em Agosto De 2008 Por UF', fontsize=20)
plt.xlabel('Unidade', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.xticks( fontsize=12, rotation=90)
plt.yticks(fontsize=12)
    
#Mostrando o gráfico
plt.show()
```

```python
dados['2020/Ago'].replace('-', np.nan, inplace=True)
```

```python
dados['2020/Ago'] = dados['2020/Ago'].astype('float64').dropna()
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(12,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando o gráfico 
sns.barplot(x='Unidade da Federação', y='2020/Ago', data=dados.sort_values('2020/Ago').dropna(), palette='Purples', ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title('Gasto Hospitalar Público Processado Em Agosto De 2020 Por UF', fontsize=20)
plt.xlabel('Unidade Federativa', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.xticks( fontsize=12, rotation=90)
plt.yticks(fontsize=12)
    
#Mostrando o gráfico
plt.show()
```
