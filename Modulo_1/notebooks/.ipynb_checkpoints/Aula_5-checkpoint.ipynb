```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv("C:/Users/pedro/Documents/Bootcamp_alura/Modulo_1/dados/A083728189_28_143_208_atendimento.csv", encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",", engine="python")
```

```python
pd.options.display.float_format = "{:.2f}".format
dados.head()
```

```python
col_preenchidas = dados.mean().index
col_preenchidas = col_preenchidas.insert(0, 'Unidade da Federação')
filtrados = dados[col_preenchidas]
filtrados.head()
```

```python
filtrados.set_index('Unidade da Federação', inplace=True)
```

```python
filtrados.drop('Total', axis=1, inplace=True)
filtrados.head()
```

```python
col_regiao=[]
regioes=['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']

for row in filtrados.iterrows():
    
    estado=row[0]
    num_reg = int(estado.split(' ')[0])
    new_estado = estado.split(' ', maxsplit =1)[1]
    filtrados = filtrados.rename(index={estado:new_estado})
    
    if num_reg < 20:
        col_regiao.append(regioes[0])
    elif num_reg < 30:
        col_regiao.append(regioes[1])
    elif num_reg < 40:
        col_regiao.append(regioes[2])
    elif num_reg < 50:
        col_regiao.append(regioes[3])
    else:
        col_regiao.append(regioes[4])
```

```python
filtrados['regiao']=col_regiao
filtrados.head()
```

```python
filtrados['Total'] = filtrados.sum(axis=1)
filtrados.head()
```

```python
ordenado = filtrados.sort_values('Total', ascending=False).drop('Total', axis=1)
```

```python
ordenado = ordenado.iloc[:,6:]
ordenado.head()
```

# Desafio 01: Refaça as manipulações feitas em aulas e gaste um tempo analisando os dados e entendendo os efeitos das manipulações

```python
lista=[]
for data in ordenado.drop('regiao', axis=1).columns:
    mes = data[-3:]
    if mes=='Jan' or mes=='Jul':
        lista.append(data)
    else:
        lista.append('')
```

```python
len(lista)
```

```python
estados = ordenado.head(5).index.to_list()
estados.append('Bahia')
estados
```

```python
@ticker.FuncFormatter
def million_formatter(x, pos):
    return "%.1f M" % (x/1E6)
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(12,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando uma palheta de 6 cores
sns.set_palette(sns.color_palette('Set1', 6))

#Criando o gráfico 
ordenado.drop('regiao', axis=1).loc[estados].T.plot(ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Público De Atendimentos Entre Fev De 2008 E Mar De 2021 Por UF', fontsize=20)
plt.xlabel('Ano/Mês', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(range(len(lista)), lista, fontsize=12, rotation=90)
plt.legend(title='UF', bbox_to_anchor=(1.05, 1), loc='upper left')

#Mostrando o gráfico
plt.show()
```

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(8,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando uma palheta de 6 cores
sns.set_palette(sns.color_palette('Set1', 6))

#Criando o gráfico 
ordenado.drop('regiao', axis=1).loc[estados].T.plot(ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Público De Atendimentos Entre Fev De 2008 E Mar De 2021 Por UF', fontsize=20)
plt.xlabel('Ano/Mês', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(range(len(lista)), lista, fontsize=12, rotation=90)
plt.legend(title='UF', bbox_to_anchor=(1.05, 1), loc='upper left')

#Mostrando o gráfico
plt.show()
```

Ao plotar o mesmo gráfico mudando apenas um parâmetro relacionado à largura da figura plotado obrservamos uma diferença visual gritante, olhando esse gráfico agora parece, em um primeiro momento, que o aumento nos gastos aconteceu de forma muito mais rápida poisa inclinação das linhas ficaram muito mais acentuadas uma vez que o espaço para plota-las diminuiu.

Podemos mudar ainda mais o gráfico para criar ilusões(o que devemos tomar cuidado para evitar na verdade).

Para criar o efeito contrário podemos aumentar a largura

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando uma palheta de 6 cores
sns.set_palette(sns.color_palette('Set1', 6))

#Criando o gráfico 
ordenado.drop('regiao', axis=1).loc[estados].T.plot(ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Público De Atendimentos Entre Fev De 2008 E Mar De 2021 Por UF', fontsize=20)
plt.xlabel('Ano/Mês', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(range(len(lista)), lista, fontsize=12, rotation=90)
plt.legend(title='UF', bbox_to_anchor=(1.05, 1), loc='upper left')

#Mostrando o gráfico
plt.show()
```

Agora aparenta que os valores aumentaram de forma bem mais lenta

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(12,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando uma palheta de 6 cores
sns.set_palette(sns.color_palette('Set1', 6))

#Criando o gráfico 
ordenado.drop('regiao', axis=1).loc[estados].T.plot(ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Público De Atendimentos Entre Fev De 2008 E Mar De 2021 Por UF', fontsize=20)
plt.xlabel('Ano/Mês', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(range(len(lista)), lista, fontsize=12, rotation=90)
plt.legend(title='UF', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.ylim(0,3000000000)

#Mostrando o gráfico
plt.show()
```

Dessa vez aumentando o eixo y vemos que os valores parecem que estão muito mais próximos do que realmente estão, o gráfico fica distorcido e passa uma false informação, nós conseguimos esse efeito aumento o valor máximo do y, e para conseguirmos o efeito contrário devemos aumentar o valor minímo do y

```python
#Criando figura e eixos
fig, ax = plt.subplots(figsize=(12,8)) 

#Mudando o estilo de fundo
sns.set_style('darkgrid')

#Criando uma palheta de 6 cores
sns.set_palette(sns.color_palette('Set1', 6))

#Criando o gráfico 
ordenado.drop('regiao', axis=1).loc[estados].T.plot(ax=ax)

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Público De Atendimentos Entre Fev De 2008 E Mar De 2021 Por UF', fontsize=20)
plt.xlabel('Ano/Mês', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(range(len(lista)), lista, fontsize=12, rotation=90)
plt.legend(title='UF', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.ylim(100000000,350000000)

#Mostrando o gráfico
plt.show()
```

Podemos ver que limitando o valor minimo do y ate escondemos valores de outro estados, dando a entender que eles gastam muito menos do que realmente gastam.

# Desafio 02: Reflita sobre os problemas na manipulação de dados

Um dos maiores problemas na manipulação de dados é o viés, um gráfico é responsável por passar uma informação clara e objetiva para o usuário, porém ele é facilmente manipulado e distorcido afim de causar uma impressão errada, ao olhar para um gráfico o ser humano percebe os padrões muitas vezes pelo seu visual, ou seja, o visual de um gráfico é o mais importante para trasmitir a mensagem que queremos passar, porém não podemos usar desse artifício para querer passar uma descoberta 'extraordinaria', se não é a verdade, o papel do cientista de dados na visualização de dados é resolver problemas e não impressionar, a verdade muitas vezes não é da forma que imaginamos e provar uma hipótese é uma tarefa muito mais dificil que parece, por esta razão devemos passar somente a verdade e não tentar provar uma hipótese manipulando o gráfico de forma tendenciosa.
